```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Convert release_date to datetime values

```python
def dt_convert(dt):
    if dt is None:
        return None
    try:
        return pd.to_datetime(dt)
    except:
        return None

# Drop any dates that can't be converted.
sdata["release_date"] = sdata.release_date.apply(dt_convert)
```

## Process min_memory into interpretable RAM

```python
import re
```

```python
def mem_extract(string):
    if pd.isna(string):
        return None
    ram = re.findall(r"(\d+\+?\s?.{1}B)", string)
    if len(ram) == 0:
        return None
    return mem_convert(ram[0])

def mem_convert(ram):
    """Convert RAM measurements to Gigabytes"""
    convert = {
        "GB":1,
        "MB":0.001
    }
    measure = ram[-2:]
    num = int(re.findall(r"\d+", ram)[0])
    return num * convert[measure]
```

```python
sdata["min_memory"] = sdata["min_memory"].apply(mem_extract)
sdata["min_memory"].plot(kind="hist")
```

## Process min_storage into interpretable sizes

```python
sdata["min_storage"] = sdata["min_storage"].apply(mem_extract)
```

## Process min_processor into Ghz

```python
cpus = pd.read_csv(ROOT_DIR + "data/Intel_CPUs.csv")
cpus = cpus.dropna(subset=["Processor_Number"])
```

```python
cpus.shape
```

```python
def hrtz_convert(hrtz):
    if pd.isna(hrtz):
        return None
    measure = re.findall(r"\+?\s?(\D+)", hrtz)[-1][0].lower()
    num = float(re.findall(r"(\d\.?\d*)", hrtz)[0])
    if measure == 'g':
        return num
    elif measure == 'm':
        return num * 0.001

def proc_num_lookup(proc_num):
    cpu = cpus[cpus["Processor_Number"].str.contains(proc_num)]
    if cpu.shape[0] == 0:
        return None
    cpu = cpu.iloc[0].dropna()
    if "Max_Turbo_Frequency" in cpu.index:
        return cpu["Max_Turbo_Frequency"]
    elif "Processor_Base_Frequency" in cpu.index:
        return cpu["Processor_Base_Frequency"]
    return None

def get_cpu_speed(cpu_str):
    if pd.isna(cpu_str):
        return None
    
    cpu_speed = re.findall(r"(\d+\.?,?\d*\+?\s?(?i:[g|m]hz))\s?", cpu_str)
    if len(cpu_speed) > 0:
        cpu_speed = cpu_speed[0]
    else:
        proc_num = re.findall(r"(i\d?\s?-?\s?\d+\w?)[/|\s|]?", cpu_str)
        if len(proc_num) == 0:
            proc_num = re.findall(r"(\w\d+)\s", cpu_str)
            if len(proc_num) == 0:
                return None
        proc_num = proc_num[0].replace('k', 'K').replace(' ', '-')
        proc_num = re.sub(r'-+', '-', proc_num)
        cpu_speed = proc_num_lookup(proc_num)
        if cpu_speed is None:
            return cpu_speed
    
    cpu_speed = cpu_speed.replace(',', '.')
    cpu_speed = hrtz_convert(cpu_speed)
    
    return cpu_speed
```

```python
sdata["min_processor"] = sdata["min_processor"].apply(get_cpu_speed)
```

## Looking at the distribution of game releases over time

```python
date_counts = sdata.release_date.dt.year.value_counts().to_frame().reset_index()
date_counts = date_counts.rename({"index":"release year",
                                  "release_date":"count"}, axis=1)
# Dropping 2021 because it's an incomplete year thus far
# Not dropping it would affect fitting any models to the data
date_counts = date_counts.drop(date_counts[date_counts["release year"] == 2021].index)
```

### Fitting a polynomial line to the game count per year

```python
x_2, x, b = np.polyfit(date_counts["release year"], date_counts["count"], 2)
```

```python
pred_data = []
for year in date_counts["release year"].unique():
    pred_data.append({
        "year":year,
        "pred_count":(x_2) * (year ** 2) + (x * year) + b
    })
pred_data = pd.DataFrame(pred_data).sort_values("year")
```

```python
plt.plot(pred_data["year"], pred_data["pred_count"])
plt.scatter(x=date_counts["release year"], y=date_counts["count"])
plt.show()
```

### Log-scaled fit

```python
x_2, x, b = np.polyfit(date_counts["release year"], np.log(date_counts["count"]), 2)
```

```python
pred_data = []
for year in date_counts["release year"].unique():
    pred_data.append({
        "year":year,
        "pred_count":(x_2) * (year ** 2) + (x * year) + b
    })
pred_data = pd.DataFrame(pred_data).sort_values("year")
```

```python
plt.plot(pred_data["year"], pred_data["pred_count"])
plt.scatter(x=date_counts["release year"], y=np.log(date_counts["count"]))
plt.show()
```
