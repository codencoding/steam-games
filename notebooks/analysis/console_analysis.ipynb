```python
def gpu_conv(flops):
    if pd.isna(flops):
        return None
    data = flops.split(' ')
    if data[-1][0] == 'G':
        return float(data[0])
    else:
        return float(data[0]) * 1000

def date_conv(date):
    if pd.isna(date):
        return None
    month, day, year = date.split('/')
    month = int(month)
    day = int(day)
    year = int(year)
    return datetime.datetime(year, month, day)
```

```python
df = pd.read_csv(ROOT_DIR + "data/console_specs.csv")
df["gpu flops"] = df["gpu flops"].apply(gpu_conv)
df["release date"] = df["release date"].apply(date_conv)

xbox_df = df[df["brand"] == "xbox"].reset_index(drop=True)
ps_df = df[df["brand"] == "playstation"].reset_index(drop=True)
```

```python
brand_colors = {
    "xbox":"lightgreen",
    "playstation":"lightblue"
}
```

```python
groups = df.groupby("brand")
for name, group in groups:
    b_color = brand_colors[name]
    plt.plot(group["release date"].apply(lambda row:row.year), group["cpu freq (GHz)"], label=name, linewidth=6, alpha=0.7, color=b_color)
    plt.scatter(group["release date"].apply(lambda row:row.year), group["cpu freq (GHz)"], s=200, alpha=0.5, color=b_color)

plt.title("2013 console releases had a dip in CPU potency?")
plt.ylabel("CPU Frequency (GHz)")
plt.xlabel("Release Year")
min_year = df["release date"].min().year
max_year = df["release date"].max().year
xticks = np.linspace(min_year, max_year, int((max_year-min_year)/2)+1)
plt.xticks(xticks, rotation=45)
plt.legend()
plt.show()
```

```python
groups = df.groupby("brand")
for name, group in groups:
    b_color = brand_colors[name]
    plt.plot(group["release date"].apply(lambda row:row.year), group["ram (GB)"], label=name, linewidth=6, alpha=0.7, color=b_color)
    plt.scatter(group["release date"].apply(lambda row:row.year), group["ram (GB)"], s=200, alpha=0.5, color=b_color)

plt.title("Increasing RAM specs over time")
plt.ylabel("RAM (GB)")
plt.xlabel("Release Year")
min_year = df["release date"].min().year
max_year = df["release date"].max().year
xticks = np.linspace(min_year, max_year, int((max_year-min_year)/2)+1)
plt.xticks(xticks, rotation=45)
plt.legend()
plt.show()
```
