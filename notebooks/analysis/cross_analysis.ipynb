## Load steam hardware data

```python
# Load system requirements data
data = pd.read_csv(ROOT_DIR + "data/sys_reqs/1-19-2021.sys_reqs.csv")
# Load cpu hardware specs data
cpus = pd.read_csv(ROOT_DIR + "data/Intel_CPUs.csv")
cpus = cpus.dropna(subset=["Processor_Number"])
# Preprocess the system reqs data
steam_df = pr.preprocess_data(data, cpus)
steam_df = pr.filter_data(steam_df)
# Cutoff release year at 2021
steam_df = steam_df[steam_df.release_date.dt.year < 2022].reset_index(drop=True)
```

## Load console data

```python
console_df = pd.read_csv(ROOT_DIR + "data/console_specs.csv")
console_df["gpu flops"] = console_df["gpu flops"].apply(pr.gpu_conv)
console_df["release date"] = console_df["release date"].apply(pr.date_conv)
```

```python
brand_colors = {
    "xbox":"lightgreen",
    "playstation":"lightblue"
}
```

```python
groups = console_df.groupby("brand")
for name, group in groups:
    b_color = brand_colors[name]
    plt.plot(group["release date"], group["cpu freq (GHz)"], label=name, linewidth=6, alpha=1, color=b_color)
    plt.scatter(group["release date"], group["cpu freq (GHz)"], s=200, alpha=1, color=b_color)

plt.scatter(steam_df[("release_date")], steam_df[("min_processor")], alpha=0.2, color="lightsalmon", zorder=-1, label="steam games")
plt.title("2013 console releases had a dip in CPU potency?")
plt.ylabel("CPU Frequency (GHz)")
plt.xlabel("Release Year")
min_year = console_df["release date"].min().year
max_year = console_df["release date"].max().year
xticks = np.linspace(min_year, max_year, int((max_year-min_year)/2)+1)
# plt.xticks(xticks, rotation=45)
plt.legend()
plt.show()
```

```python
groups = console_df.groupby("brand")
for name, group in groups:
    b_color = brand_colors[name]
    plt.plot(group["release date"], group["ram (GB)"], label=name+" specifications", linewidth=6, alpha=1, color=b_color)
    plt.scatter(group["release date"], group["ram (GB)"], s=200, alpha=1, color=b_color)

plt.scatter(steam_df[("release_date")], steam_df[("min_memory")], alpha=0.2, color="lightsalmon", zorder=-1, label="steam game requirements")
plt.title("Although RAM size is being increasingly accommodated, what about RAM speed?")
plt.ylabel("RAM (GB)")
plt.xlabel("Release Year")
min_year = console_df["release date"].min().year
max_year = console_df["release date"].max().year
xticks = np.linspace(min_year, max_year, int((max_year-min_year)/2)+1)
# plt.xticks(xticks, rotation=45)
plt.legend()
plt.show()
```

```python
def get_latest_console(row, console):
    if pd.isnull(row.release_date):
        return None
    game_year = row.release_date.year
    
    latest_console = console_df[console_df.brand==console]
    if game_year < latest_console["release date"].dt.year.min():
        return None
    latest_console = latest_console[latest_console["release date"].dt.year <= game_year].sort_values("release date")
    
    return latest_console.iloc[-1].console
```

```python
console = "playstation"
# console = "xbox"


biggest_games = []

grouped_years = steam_df["release_date"].dt.to_period("Y")
for year, group_df in steam_df.groupby(grouped_years):
    biggest_game = group_df.sort_values("min_storage", ascending=False).iloc[0]
    biggest_game["latest_console"] = get_latest_console(biggest_game, console)
    biggest_games.append(biggest_game)
    
biggest_games = pd.DataFrame(biggest_games)
biggest_games = biggest_games.reset_index(drop=True)

# Inner merge to exclude games before the release of the first console
# and to exclude the currently untargeted console.
merged_data = biggest_games.merge(console_df, "inner", left_on="latest_console", right_on="console")

console_value = merged_data["min_storage"] / merged_data["storage (GB)"]
console_value.index = merged_data.release_date.dt.year
console_value = console_value.replace(np.inf, np.nan).dropna()
```
